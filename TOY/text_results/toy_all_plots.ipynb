```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import numpy as np

# Labels for the plots
labels = [
    'FCCM', r'$\mu$-BALD', r'$\rho$-BALD', r'$\mu\rho$-BALD', 'QHTE', 'MACAL',
    'LCMD', 'BADGE', 'BAIT'
]

# Colors for the plots
colours = [
    'red', 'blue', 'orange', 'green', 'black', 'black',
    'brown', 'pink', 'gray'
]

plt.figure(figsize=(6, 5))  # Width = 10, Height = 6 (adjust as needed)

# Define the selected x-axis points
selected_x = [1, 10, 20, 30, 40, 50]

# Loop through each files_path and label
for index, files_path in enumerate([
    "fccm/*.csv", "mu/*.csv", "rho/*.csv", "murho/*.csv", "qhte/*.csv", "qhte/*.csv",
    "lcmd/*.csv", "kmeanspp/*.csv", "bait/*.csv"
]):
    # List to store dataframes for each file
    dfs = []

    try:
        # Read each CSV file and append it to the list
        for file in glob.glob(files_path):
            df = pd.read_csv(file, index_col=0)
            dfs.append(df.loc["PEHE"])  # Extract the row corresponding to "PEHE" for each file
    except pd.errors.EmptyDataError:
        # Skip this iteration if there are no files in the directory
        continue

    # Check if there are no files in the directory
    if not dfs:
        continue

    # Concatenate all dataframes into a single dataframe
    combined_df = pd.concat(dfs, axis=1)
    
    # Calculate the mean for each "number of samples"
    mean_values = combined_df.mean(axis=1)
    
    # Calculate the variance for each "number of samples"
    std_values = combined_df.std(axis=1)

    # Convert the index to integers to match selected_x
    mean_values.index = mean_values.index.astype(int)
    std_values.index = std_values.index.astype(int)

    # Filter only the rows that match selected_x
    mean_series = mean_values.loc[mean_values.index.intersection(selected_x)]
    std_series = std_values.loc[std_values.index.intersection(selected_x)]

    # Plot mean values
    plt.plot(mean_series.index, mean_series.values, marker='o', markersize=5, linestyle='-', label=f"{labels[index]}", color=colours[index])

    # Add error bars for confidence intervals (I-shape)
    plt.errorbar(
        mean_series.index,        # x values
        mean_series.values,       # y values (mean)
        yerr=std_series.values,   # Confidence intervals (std deviation)
        fmt='o',                  # Marker for the mean point
        color=colours[index],     # Color for the line and markers
        capsize=5,                # Adds horizontal caps to the error bars
        alpha=0.6                 # Adjust transparency if needed
    )


# Add labels and title
plt.xlabel("Consumption of the Budget")
plt.ylabel("$\sqrt{\epsilon_{PEHE}}$")

# Define the corresponding labels
x_labels = ['0%', '20%', '40%', '60%', '80%', '100%']

# Set the x-ticks and their labels
plt.xticks(ticks=selected_x, labels=x_labels)


# Adjust legend to display horizontally and set the position
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.25), ncol=3,frameon=False)

# Add grid lines for better readability
plt.grid(visible=True, linestyle='--', alpha=0.7)

# Set the title for the plot
#plt.title("PEHE Analysis with Selected Points and Confidence Intervals")

# Show the plot
plt.tight_layout()
# Save the figure
plt.savefig('TOY_PEHE.pdf', bbox_inches='tight')  # Save without extra margins

plt.show()
```
